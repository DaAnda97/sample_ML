```python
import pandas as pd
import numpy as np
np.random.seed(42)

import seaborn as sns
```

## Look at the distribution of the data

```python
df_train = pd.read_csv('data/genosharing-train.csv')
df_train.head()
```

```python
print(df_train['OUT'].value_counts())

sns.barplot(x=df_train.OUT.value_counts().index, y=df_train.OUT.value_counts())
```

```python
sns.pairplot(df_train, hue="OUT", vars=["IN-1", "IN-2", "IN-3", "IN-4"])
```

```python
# Divide by class
df_class_0 = df_train[df_train['OUT'] == 0]
df_class_1 = df_train[df_train['OUT'] == 1]
df_class_n1 = df_train[df_train['OUT'] == -1]

# Class count rows
count_class_0 = df_class_0.shape[0]
count_class_1 = df_class_1.shape[0]
count_class_n1 = df_class_n1.shape[0]

print(" 0: {} \n 1: {} \n-1: {}".format(count_class_0, count_class_1, count_class_n1))
```

## Oversampling with pandas

```python
# sample up to amount of class_0
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
df_class_n1_over = df_class_n1.sample(count_class_0, replace=True)

# concatenate
df_train_over = pd.concat([df_class_0, df_class_1_over, df_class_n1_over], axis=0, ignore_index=True)

print('Random over-sampling:')
print(df_train_over.OUT.value_counts())

sns.barplot(x=df_train_over.OUT.value_counts().index, y=df_train_over.OUT.value_counts())
```

```python
df_train_over.describe()
```

```python
sns.pairplot(df_train_over, hue="OUT", vars=["IN-1", "IN-2", "IN-3", "IN-4"])
```
